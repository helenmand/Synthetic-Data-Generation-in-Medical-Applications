```python
import pandas as pd
from imblearn.over_sampling import SMOTE
```

```python
target_variable = 'Diagnosis' #'num'
real_data = pd.read_csv('datasets/breast.csv')
X_train = real_data.drop(columns=[target_variable])
y_train = real_data[target_variable]
```

```python
smote_model = SMOTE(random_state=42, k_neighbors=5)
X_synthetic, y_synthetic = smote_model.fit_resample(X_train, y_train)
X_synthetic
```

```python
y_synthetic.value_counts()
```

```python
pd.concat([X_synthetic, y_synthetic], axis=1).to_csv('synthetic/smote.csv', index=False)
```
