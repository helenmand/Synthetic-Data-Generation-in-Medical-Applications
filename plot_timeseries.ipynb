```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Real time series

```python
file_path = 'datasets/baseline.csv'
df = pd.read_csv(file_path)
features = ['acc_var', 'hr', 'rmssd', 'sdnn', 'temp', 'eda', 'eda_freq', 'bm', 'location', 'concentration']
```

```python
# Plot all users per each feature
for feature in features:
    plt.figure(figsize=(12, 6))
    for user in df['user_id'].unique():
        user_data = df[df['user_id'] == user]
        plt.plot(user_data['time'], user_data[feature], label=f'User {user}')
    plt.xlabel('Time')
    plt.ylabel(feature)
    plt.title(f'{feature} over Time for All Users')
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Synthetic time series

```python
file_path = 'synthetic/synthetic_data_timeseries1.csv'
synthetic_data = pd.read_csv(file_path)
```

```python
synthetic_data['time'] = pd.to_datetime(synthetic_data['time'], unit='s')  # Assuming the integer time is in seconds

# List of features to plot
features = ['acc_var', 'hr', 'rmssd', 'sdnn', 'temp', 'eda', 'eda_freq', 'bm', 'location', 'concentration']

# Plot all users per each feature
for feature in features:
    plt.figure(figsize=(12, 6))
    for user in synthetic_data['user_id'].unique():
        user_data = synthetic_data[synthetic_data['user_id'] == user]
        plt.plot(user_data['time'], user_data[feature], label=f'User {user}')
    plt.xlabel('Time')
    plt.ylabel(feature)
    plt.title(f'{feature} over Time for All Users')
    plt.legend()
    plt.grid(True)
    plt.show()
```
