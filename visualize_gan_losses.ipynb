```python
import matplotlib.pyplot as plt

# Define the path to your loss file
file_path = 'wgan_loss.txt'

# Initialize lists to hold the values
epochs = []
batches = []
d_losses = []
g_losses = []

# Read the file
with open(file_path, 'r') as file:
    for line in file:
        # Split the line into components
        parts = line.strip().split(' ')
        
        # Extract the epoch, batch, D loss, and G loss values
        epoch = int(parts[1].split('/')[0].strip('[]'))
        batch = int(parts[3].split('/')[0].strip('[]'))
        d_loss = float(parts[6].strip('[]'))
        g_loss = float(parts[9].strip('[]'))
        
        # Append to the lists
        epochs.append(epoch)
        batches.append(batch)
        d_losses.append(d_loss)
        g_losses.append(g_loss)

# Plot the losses
plt.figure(figsize=(14, 7))

# Plot D loss and G loss on the same plot
plt.plot(d_losses, label='D loss')
plt.plot(g_losses, label='G loss', color='orange')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('WGAN Discriminator and Generator Loss for KneeXray Dataset')
plt.legend()

plt.tight_layout()
plt.show()
```
